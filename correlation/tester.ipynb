```python
from torchvision.models.detection import fasterrcnn_resnet50_fpn_v2, FasterRCNN_ResNet50_FPN_V2_Weights
from torchvision.models.detection import maskrcnn_resnet50_fpn_v2, MaskRCNN_ResNet50_FPN_V2_Weights
weights = FasterRCNN_ResNet50_FPN_V2_Weights.COCO_V1
weights2 = MaskRCNN_ResNet50_FPN_V2_Weights.COCO_V1


preprocess = weights.transforms()
preprocess2 = weights2.transforms()

model = fasterrcnn_resnet50_fpn_v2(weights=weights)
model2 = maskrcnn_resnet50_fpn_v2(weights=weights2)

from torchvision.io import read_image
img = read_image('/rhome/pranavk/multitask-learning/data/test2017/000000000001.jpg')
batch = [preprocess(img)]
batch2 = [preprocess2(img)]
model.eval()
model2.eval()
print(model2(batch2))
```
